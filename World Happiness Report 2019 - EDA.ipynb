```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df.keys()
```

```python
X=df[['Ladder', 'SD of Ladder', 'Positive affect',
       'Negative affect', 'Social support', 'Freedom', 'Corruption',
       'Generosity', 'Log of GDP\nper capita']]
y=df['Healthy life\nexpectancy']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
linreg=LinearRegression()
```

```python
linreg.fit(X_train,y_train)
```

```python
pred=linreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,pred))
print('Mean Squarred Error:',metrics.mean_squared_error(y_test,pred))
print('Rooted Mean Squarred Error:',np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
plt.scatter(y_test,pred)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=0.01)
```

```python
ridge.fit(X_train,y_train)
```

```python
pred=ridge.predict(X_train)
```

```python
print('Rooted Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_train,pred)))
```

```python
pred_test_ridge=ridge.predict(X_test)
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,pred_test_ridge))
print('MSE:',metrics.mean_squared_error(y_test,pred_test_ridge))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,pred_test_ridge)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(n_estimators=1000)
```

```python
rfr.fit(X_train,y_train)
```

```python
pred=rfr.predict(X_test)
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,pred))
print('MSE:',metrics.mean_squared_error(y_test,pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,pred)))
```
